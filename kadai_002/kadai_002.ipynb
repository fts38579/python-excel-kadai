```python
# ============================================================
# æ¥­ç¸¾ãƒ‡ãƒ¼ã‚¿åˆ†æãƒ—ãƒ­ã‚°ãƒ©ãƒ 
# Google Colaboratoryç”¨
# ============================================================

# å¿…è¦ãªãƒ©ã‚¤ãƒ–ãƒ©ãƒªã®ã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ«
print("=" * 60)
print("ãƒ©ã‚¤ãƒ–ãƒ©ãƒªã‚’ã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ«ä¸­...")
print("=" * 60)
!pip install -q pandas openpyxl

import pandas as pd
from google.colab import files

print("âœ“ ã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ«å®Œäº†\n")

def performance(row, avg_sales):
    """
    æ¥­ç¸¾ãƒ©ãƒ³ã‚¯ã‚’åˆ¤å®šã™ã‚‹é–¢æ•°

    Args:
        row: ãƒ‡ãƒ¼ã‚¿ãƒ•ãƒ¬ãƒ¼ãƒ ã®è¡Œãƒ‡ãƒ¼ã‚¿
        avg_sales: å…¨ä½“ã®å¹³å‡å£²ä¸Š

    Returns:
        str: æ¥­ç¸¾ãƒ©ãƒ³ã‚¯ï¼ˆA, B, Cï¼‰
    """
    sales = row['å£²ä¸Š']

    if sales >= avg_sales + 50:
        return 'A'
    elif sales >= avg_sales:
        return 'B'
    else:
        return 'C'

def create_performance_report():
    """
    æ¥­ç¸¾ãƒ¬ãƒãƒ¼ãƒˆã‚’ä½œæˆã™ã‚‹é–¢æ•°
    """
    print("=" * 60)
    print("ğŸ“Š æ¥­ç¸¾ãƒ‡ãƒ¼ã‚¿ã‚’åˆ†æä¸­...")
    print("=" * 60)

    # === Step 1: ãƒ‡ãƒ¼ã‚¿ãƒ•ãƒ¬ãƒ¼ãƒ ã®ä½œæˆ ===
    print("\nâœ“ Step 1: ãƒ‡ãƒ¼ã‚¿ãƒ•ãƒ¬ãƒ¼ãƒ ã‚’ä½œæˆ")

    data = {
        'æ—¥ä»˜': ['2023-05-17', '2023-05-18', '2023-05-19', '2023-05-20', '2023-05-21'],
        'ç¤¾å“¡å': ['å±±ç”°', 'ä½è—¤', 'éˆ´æœ¨', 'ç”°ä¸­', 'é«˜æ©‹'],
        'å£²ä¸Š': [100, 200, 150, 300, 250],
        'éƒ¨é–€': ['ãƒ¡ãƒ¼ã‚«ãƒ¼', 'ä»£ç†åº—', 'ãƒ¡ãƒ¼ã‚«ãƒ¼', 'å•†ç¤¾', 'ä»£ç†åº—']
    }

    df = pd.DataFrame(data)

    print("\nã€å…ƒãƒ‡ãƒ¼ã‚¿ã€‘")
    print(df)

    # === Step 2: å¹³å‡å£²ä¸Šã®ç®—å‡º ===
    print("\nâœ“ Step 2: å¹³å‡å£²ä¸Šã‚’ç®—å‡º")

    avg_sales = df['å£²ä¸Š'].mean()
    print(f"\nå…¨ä½“ã®å¹³å‡å£²ä¸Š: {avg_sales}")

    # å¹³å‡å£²ä¸Šåˆ—ã‚’è¿½åŠ ï¼ˆå…¨è¡Œã«åŒã˜å€¤ï¼‰
    df['å¹³å‡å£²ä¸Š'] = avg_sales

    # === Step 3: æ¥­ç¸¾ãƒ©ãƒ³ã‚¯ã®è¨­å®š ===
    print("\nâœ“ Step 3: æ¥­ç¸¾ãƒ©ãƒ³ã‚¯ã‚’è¨­å®š")

    # performanceé–¢æ•°ã‚’ä½¿ã£ã¦æ¥­ç¸¾ãƒ©ãƒ³ã‚¯ã‚’è¨­å®š
    df['æ¥­ç¸¾ãƒ©ãƒ³ã‚¯'] = df.apply(lambda row: performance(row, avg_sales), axis=1)

    print("\nã€æ¥­ç¸¾ãƒ©ãƒ³ã‚¯è¨­å®šå¾Œã€‘")
    print(df)

    # === Step 4: Excelãƒ•ã‚¡ã‚¤ãƒ«ã«æ›¸ãå‡ºã— ===
    print("\nâœ“ Step 4: Excelãƒ•ã‚¡ã‚¤ãƒ«ã«æ›¸ãå‡ºã—")

    filename = 'æ¥­ç¸¾.xlsx'
    df.to_excel(filename, index=False, sheet_name='æ¥­ç¸¾ãƒ‡ãƒ¼ã‚¿')

    print(f"\nâœ“ Excelãƒ•ã‚¡ã‚¤ãƒ«ã‚’ä½œæˆã—ã¾ã—ãŸ: {filename}")

    # === çµæœã®ã‚µãƒãƒªãƒ¼è¡¨ç¤º ===
    print("\n" + "=" * 60)
    print("ğŸ“ˆ æ¥­ç¸¾ã‚µãƒãƒªãƒ¼")
    print("=" * 60)
    print(f"\nå…¨ä½“ã®å¹³å‡å£²ä¸Š: {avg_sales}")
    print(f"\næ¥­ç¸¾ãƒ©ãƒ³ã‚¯åˆ¥ã®äººæ•°:")
    print(df['æ¥­ç¸¾ãƒ©ãƒ³ã‚¯'].value_counts().sort_index())

    print("\nå„ç¤¾å“¡ã®æ¥­ç¸¾:")
    for _, row in df.iterrows():
        rank = row['æ¥­ç¸¾ãƒ©ãƒ³ã‚¯']
        print(f"  {row['ç¤¾å“¡å']}: å£²ä¸Š{row['å£²ä¸Š']} â†’ ãƒ©ãƒ³ã‚¯{rank}")

    return filename

# å®Ÿè¡Œ
if __name__ == "__main__":
    print("=" * 60)
    print("ğŸš€ æ¥­ç¸¾ãƒ‡ãƒ¼ã‚¿åˆ†æé–‹å§‹")
    print("=" * 60)

    filename = create_performance_report()

    print("\n" + "=" * 60)
    print("ğŸ“¥ ãƒ€ã‚¦ãƒ³ãƒ­ãƒ¼ãƒ‰ã‚’é–‹å§‹ã—ã¾ã™...")
    print("=" * 60)

    # è‡ªå‹•ãƒ€ã‚¦ãƒ³ãƒ­ãƒ¼ãƒ‰
    files.download(filename)

    print("\nâœ… å®Œäº†ï¼")
    print(f"ğŸ’¡ ãƒ•ã‚¡ã‚¤ãƒ«å: {filename}")
    print("ğŸ“ ãƒ–ãƒ©ã‚¦ã‚¶ã®ãƒ€ã‚¦ãƒ³ãƒ­ãƒ¼ãƒ‰ãƒ•ã‚©ãƒ«ãƒ€ã‚’ç¢ºèªã—ã¦ãã ã•ã„")
```
