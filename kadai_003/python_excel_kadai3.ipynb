```python
# ============================================================
# Step3: ãƒ‡ãƒ¼ã‚¿ã®é€£çµã¨é›†è¨ˆ
# ============================================================

import pandas as pd

print("=" * 60)
print("ğŸ“Š Step3: ãƒ‡ãƒ¼ã‚¿ã®é€£çµã¨é›†è¨ˆ")
print("=" * 60)

# === 1. Excelãƒ•ã‚¡ã‚¤ãƒ«ã®èª­ã¿è¾¼ã¿ ===
print("\nã€1ã€‘Excelãƒ•ã‚¡ã‚¤ãƒ«ã‚’èª­ã¿è¾¼ã¿ä¸­...")

df_2022 = pd.read_excel('2022_å¹´é–“å£²ä¸Šè¡¨.xlsx')
print(f"âœ“ 2022å¹´ãƒ‡ãƒ¼ã‚¿èª­ã¿è¾¼ã¿å®Œäº†: {len(df_2022)}ä»¶")
print(df_2022.head())

df_2023 = pd.read_excel('2023_å¹´é–“å£²ä¸Šè¡¨.xlsx')
print(f"\nâœ“ 2023å¹´ãƒ‡ãƒ¼ã‚¿èª­ã¿è¾¼ã¿å®Œäº†: {len(df_2023)}ä»¶")
print(df_2023.head())

# === 2. ãƒ‡ãƒ¼ã‚¿ã®é€£çµ ===
print("\n" + "=" * 60)
print("ã€2ã€‘ãƒ‡ãƒ¼ã‚¿ã‚’é€£çµä¸­...")
print("=" * 60)

df_combined = pd.concat([df_2022, df_2023], ignore_index=True)
print(f"\nâœ“ é€£çµå®Œäº†: {len(df_combined)}ä»¶")
print(f"  (2022å¹´: {len(df_2022)}ä»¶ + 2023å¹´: {len(df_2023)}ä»¶)")

print("\nã€é€£çµå¾Œã®ãƒ‡ãƒ¼ã‚¿ã€‘")
print(df_combined)

# === 3. ã‚°ãƒ«ãƒ¼ãƒ—åŒ–ã¨é›†è¨ˆ ===
print("\n" + "=" * 60)
print("ã€3ã€‘å•†å“ãƒ»å£²ä¸Šå¹´ã”ã¨ã«ã‚°ãƒ«ãƒ¼ãƒ—åŒ–ã—ã¦é›†è¨ˆä¸­...")
print("=" * 60)

df_grouped = df_combined.groupby(['å•†å“', 'å£²ä¸Šå¹´'])['é‡‘é¡ï¼ˆåƒå††ï¼‰'].sum().reset_index()
print(f"\nâœ“ é›†è¨ˆå®Œäº†: {len(df_grouped)}ä»¶")

print("\nã€é›†è¨ˆçµæœã€‘")
print(df_grouped)

# === 4. ã‚µãƒãƒªãƒ¼æƒ…å ± ===
print("\n" + "=" * 60)
print("ğŸ“ˆ é›†è¨ˆã‚µãƒãƒªãƒ¼")
print("=" * 60)

print(f"\nå•†å“ç¨®é¡æ•°: {df_grouped['å•†å“'].nunique()}ç¨®é¡")
print(f"å¯¾è±¡å¹´æ•°: {df_grouped['å£²ä¸Šå¹´'].nunique()}å¹´")

print("\nã€å•†å“åˆ¥ã®åˆè¨ˆï¼ˆå…¨å¹´åˆè¨ˆï¼‰ã€‘")
product_total = df_combined.groupby('å•†å“')['é‡‘é¡ï¼ˆåƒå††ï¼‰'].sum().sort_values(ascending=False)
print(product_total)

print("\nã€å¹´åˆ¥ã®åˆè¨ˆã€‘")
year_total = df_combined.groupby('å£²ä¸Šå¹´')['é‡‘é¡ï¼ˆåƒå††ï¼‰'].sum()
print(year_total)

print("\n" + "=" * 60)
print("âœ… Step3 å®Œäº†ï¼")
print("=" * 60)
print("\nğŸ’¡ æ¬¡ã¯Step4ã§Excelãƒ•ã‚¡ã‚¤ãƒ«ã«æ›¸ãè¾¼ã¿ã¾ã™")

# ã‚°ãƒ«ãƒ¼ãƒ—åŒ–ã—ãŸãƒ‡ãƒ¼ã‚¿ã‚’æ¬¡ã®ã‚¹ãƒ†ãƒƒãƒ—ã§ä½¿ã†ãŸã‚ã«ä¿å­˜
df_result = df_grouped
print(f"\nğŸ“Œ é›†è¨ˆçµæœã‚’å¤‰æ•° df_result ã«ä¿å­˜ã—ã¾ã—ãŸ")
```
